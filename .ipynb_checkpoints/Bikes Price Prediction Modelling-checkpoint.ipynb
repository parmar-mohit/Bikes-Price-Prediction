# Bikes Price Prediction Modelling

```python
# Importing necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
from xgboost import XGBRegressor

# importing data
df = pd.read_csv("bikes_data_cleaned.csv")
df.head()
```

```python
# Sepearting X and Y
X = df.drop("Price (INR)",axis=1)
Y = df["Price (INR)"]
```

```python
# Seperating Training and Testing Data
X_train,Y_train,X_test,Y_test = train_test_split(X,Y,test_size=0.4)
```

Since we have categorical varibles in our dataset we have 3 options to deal with them
* Drop categorical variables
* Ordinal Encoding
* One Hot Encoding

We will see how our model perform in each of theese method by using Random Forest Regressor Model

```python
categorical_variables = ["Company","Country of Origin","Looks","Body Type"]

def getMeanAbsoluteError(X_train,Y_train,X_test,Y_test):
    model = DecisionTreeRegressor()
    model.fit(X_train,Y_train)
    predictions = model.predict(X_test)
    return mean_absolute_error(Y_test,predictions)

# Dropping Categorical Variables
X_drop_train = X_train.drop(categorical_variables,axis=1)
X_drop_test = X_test.drop(categorical_variables,axis=1)

print("Mean Absolute Error after dropping categorical variables : ",getMeanAbsoluteError(X_drop_train,Y_train,X_drop_test,Y_test))

# Using Oridnal Encoding
X_ordinal_train = X_train.copy()
X_ordinal_test = X_test.copy()
ordinal_encoder = OrdinalEncoder()
X_ordinal_train[categorical_variables] = ordinal_encoder.fit_transform(X_ordinal_train[categorical_variables])
X_ordinal_test[categorical_variables] = ordinal_encoder.transform(X_ordinal_test[categorical_variables])

print("Mean Absolute Error using Ordinal Encoding : ",getMeanAbsoluteError(X_ordinal_train,Y_train,X_oridnal_test,Y_test))

# Using One Hot Encoding
X_onehot_train = X_train.copy()
X_onehot_test = X_test.copy()
one_hot_encoder = OneHotEncoder()
X_onehot_train[categorical_variables] = one_hot_encoder.fit_transform(X_onehot_train[categorical_variables])
X_onehot_test[categorical_variables] = one_hot_encoder.transform(X_onehot_test[categorical_variables])

print("Mean Absolute Error using One Hot Encoding : ",getMeanAbsoluteError(X_onehot_train,Y_train,X_onehot_test,Y_test))
```
